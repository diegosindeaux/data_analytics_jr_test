```python
import dash
from dash import dcc, html, Input, Output
import pandas as pd
import plotly.express as px
from sqlalchemy import create_engine

DATABASE_URL = "postgresql://diegosindeaux:diego123@localhost:5432/postgres"
engine = create_engine(DATABASE_URL)
```

```python
query = "SELECT idade, sexo, quantidade_alunos FROM test_analytics.escolas_alunos_2019;"
with engine.connect() as conn:
    delete_query = """
    DELETE FROM test_analytics.escolas_alunos_2019 
    WHERE situacao_escola != 'Ativa';
    """
    conn.execute(text(delete_query))
    
# Excluí todos os dados de escolas que não estão ativas.

data = pd.read_sql_query(query, engine)

result_2019 = data.groupby(['idade', 'sexo'])['quantidade_alunos'].sum().reset_index()

result = result_2019

result.rename(columns={'quantidade_alunos' : 'quantidade_alunos_2019'}, inplace=True)

result_2019.head(40)
```

```python
query = "SELECT cep, quantidade_alunos FROM test_analytics.escolas_alunos_2019;"
data = pd.read_sql_query(query, engine)

data['cep_prefixo'] = data['cep'].astype(str).str[:2].astype(int)

def mapear_regiao(cep):
    if 0 <= cep <= 19:
        return 'Centro'
    elif 20 <= cep <= 29:
        return 'Zona Norte'
    elif 30 <= cep <= 39 or 80 <= cep <= 89:
        return 'Zona Leste'
    elif 40 <= cep <= 49:
        return 'Zona Sul'
    elif 50 <= cep <= 59:
        return 'Zona Oeste'
    else:
        return 'Outra Região'

data['regiao2019'] = data['cep_prefixo'].apply(mapear_regiao)


alunos_por_regiao = data.groupby('regiao2019')['quantidade_alunos'].sum().reset_index()

alunos_por_regiao = alunos_por_regiao.sort_values(by='quantidade_alunos', ascending=False)

alunos_por_regiao.head()
```
